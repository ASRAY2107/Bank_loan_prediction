```python
x = df.drop('Personal Loan',axis =1)
y = df['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 30,random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
lr = LogisticRegression(solver = 'liblinear')
```

```python
kn = KNeighborsClassifier(n_neighbors=3,weights='distance')
nb = GaussianNB()
```

```python
#using logisticRegression to train the model
lr.fit(x_train,y_train)
```

```python
y_test_pred_lr = lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(y_test,y_test_pred_lr)
rmse = np.sqrt(mse)
```

```python
print("the mean_squared error for logistic regression is {}".format(mse))
print("the root mean squared error logistic regression is  {}".format(rmse))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_test_pred_lr,labels = [1,0])
```

```python
df_cm = pd.DataFrame(cm,index = [i for i in ['1','0']],columns = [i for i in["predict 1","predict 0"]])
sns.heatmap(df_cm,annot = True)
plt.title("confusion matrix using LogisticRegression")
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test,y_test_pred_lr,labels= [1,0])
print(report)
```
